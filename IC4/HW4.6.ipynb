```python
import polars as pl 

lazy_df = pl.scan_csv("orders.csv")

query = (
    lazy_df
    .filter(pl.col("status") == "paid")
    .select(
        "country",
        "quantity"
    )
    .group_by("country")
    .agg(pl.col("quantity").sum().alias("total_quantity"))
    .sort("total_quantity", descending=True)
)

print(query.explain())

```
