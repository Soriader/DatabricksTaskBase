```python
import pandas as pd
import pyspark as ps 

df = pd.read_csv('sample_orders.csv', parse_dates=['placed_at'])

is_paid = df.query("status == 'paid'")

parse_placed_at = pd.to_datetime(df["placed_at"], format ='%Y-%m-%d', errors='coerce')

revenue = (lambda r: r.price_each * r.quantity)


country_map = {
    "Poland": "PL",
    "Germany": "DE",
    "United Kingdom": "UK",
    "Spain": "ES",
    "France": "FR",
    "Netherlands": "NL",
    "Ireland": "IE",
    "Sweden": "SE",
    "Italy": "IT",
    "Portugal": "PT"
}

country = df["country"].map(country_map)

result = is_paid.assign(country=country, parse_placed_at=parse_placed_at, revenue=revenue).groupby(["country"]).agg(revenue=("revenue", "sum"), orders=("id", "count")).sort_values("revenue", ascending=False).reset_index()

print(result)
```
