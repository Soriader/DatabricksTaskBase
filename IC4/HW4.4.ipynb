```python
import pandas as pd
import numpy as np

df = pd.read_csv("orders_hw41.csv", parse_dates=['placed_at'])

df["revenue"] = df["quantity"] * df["price_each"]

df_time = df.set_index("placed_at").sort_index()

a = df_time["revenue"].resample("D").sum()

print("A tests")
print(type(df_time.index))
print(a.head())
print(a.index)

b = a.rolling(7, min_periods=1).mean()

print("B tests")
print(b)

month = a.index.to_period("M")
c = a.groupby(month).cumsum()

print("C tests")
print(c.head())
print(c.index)

```
