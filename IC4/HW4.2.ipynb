```python
import pandas as pd
import numpy as np

df = pd.read_csv("orders.csv")

price_each = df["price_each"].to_numpy()
quantity = df["quantity"].to_numpy()
country = df["country"].to_numpy()
revenue = price_each * quantity

a = (price_each- price_each.min()) / (price_each.max() - price_each.min())


b = (quantity - quantity.mean()) / quantity.std()

df["rev"] = revenue

country_mean = df.groupby("country")["rev"].transform("mean")

country_std = df.groupby("country")["rev"].transform("std").fillna(0)

c = np.where(
    country_std != 0,
    (df["rev"] - country_mean) / country_std, 0)

#test
print(np.isfinite(c).all())
```
