```python
import pandas as pd
import numpy as np

df = pd.read_csv("orders_hw41.csv", parse_dates=['placed_at'])

revenue = df['quantity'] * df['price_each']
df['revenue'] = revenue

month = df["placed_at"].dt.to_period("M").astype(str)
df['month'] = month

paid = df[df['status'] == 'paid']

pivot = pd.pivot_table(
    paid,
    values="revenue",
    index="country",
    columns="month",
    aggfunc="sum",
    fill_value=0
)

melted = pivot.reset_index().melt(
    id_vars="country",
    var_name="month",
    value_name="revenue"
)

total_paid = paid["revenue"].sum()
total_melted = melted["revenue"].sum()

print(f"pivot shape: {pivot.shape}")
print(f"melted shape:  {pivot.shape}")
print(f"round-trip totals: {np.isclose(total_paid, total_melted)}")

```
