```python
%sql

CREATE CATALOG IF NOT EXISTS dev;
CREATE CATALOG IF NOT EXISTS test;
CREATE CATALOG IF NOT EXISTS prod;

SHOW CATALOGS;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS dev.sales;
CREATE SCHEMA IF NOT EXISTS dev.staging;
CREATE SCHEMA IF NOT EXISTS test.sales;
CREATE SCHEMA IF NOT EXISTS prod.sales;
CREATE SCHEMA IF NOT EXISTS prod.reporting;
```

```python
%sql
SHOW GROUPS;
```

```python
%sql
GRANT USE CATALOG ON CATALOG dev TO `engineering`;
GRANT USE SCHEMA, SELECT, CREATE TABLE, MODIFY ON SCHEMA dev.sales TO `engineering`;
GRANT USE SCHEMA, SELECT, CREATE TABLE, MODIFY ON SCHEMA dev.staging TO `engineering`;

GRANT USE CATALOG ON CATALOG test TO `engineering`;
GRANT USE SCHEMA, CREATE TABLE, SELECT, MODIFY ON SCHEMA test.sales TO `engineering`;


GRANT USE CATALOG ON CATALOG prod TO `analysts`;
GRANT USE SCHEMA, SELECT ON SCHEMA prod.sales TO `analysts`;
GRANT USE SCHEMA, SELECT ON SCHEMA prod.reporting TO `analysts`;

GRANT ALL PRIVILEGES ON CATALOG dev TO `data_admins`;
GRANT ALL PRIVILEGES ON CATALOG test TO `data_admins`;
GRANT ALL PRIVILEGES ON CATALOG prod TO `data_admins`;
```
