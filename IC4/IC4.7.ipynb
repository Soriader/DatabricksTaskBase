```python
import pandas as pd
from time import perf_counter

df = pd.read_csv("orders.csv")

def revenue_row(row):
 return row["quantity"] * row["price_each"]

start = perf_counter()
df["revenue_slow"] = df.apply(revenue_row, axis=1) 
end = perf_counter()
slow_time = end - start

start = perf_counter()
df["revenue_fast"] = df["quantity"] * df["price_each"]
end = perf_counter()

print((df["revenue_slow"] == df["revenue_fast"]).all())
fast_time = end - start
speedup = slow_time / fast_time

print(f"slow apply time: {slow_time}\nfast vectorized time: {fast_time}\nspeedup: {speedup}x")
```
