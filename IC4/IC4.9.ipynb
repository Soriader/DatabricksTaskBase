```python
import pandas as pd 

df_orders = pd.read_csv('orders_sample.csv')
df_customers = pd.read_csv('customers_sample.csv')

inner_join = df_orders.merge(df_customers, on="customer_id", how="inner")
print(f"Inner join row count:{inner_join.shape[0]}")

left_join = df_orders.merge(df_customers, on="customer_id", how="left", indicator=True)
print(f"Orphan order_id:{left_join[left_join['_merge'] == 'left_only']['order_id']}")

paid_orders = inner_join[inner_join["status"] == "paid"]

revenue_paid = paid_orders["quantity"] * paid_orders["price_each"]

result = revenue_paid.groupby(paid_orders["segment"]).sum().reset_index()
display(result)
```
