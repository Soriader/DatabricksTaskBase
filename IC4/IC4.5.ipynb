```python
import pandas as pd
import numpy as np

df = pd.read_csv('sample_orders.csv', parse_dates=['placed_at'])


country_map = {
    "Poland": "PL",
    "Germany": "DE",
    "United Kingdom": "UK",
    "Spain": "ES",
    "France": "FR",
    "Netherlands": "NL",
    "Ireland": "IE",
    "Sweden": "SE",
    "Italy": "IT",
    "Portugal": "PT"
}
quantity = df["quantity"].to_numpy()
price_each = df["price_each"].to_numpy()
country_code = df["country"].map(country_map).to_numpy()

conditions = [
    (price_each>= 50) & (price_each < 150),
    price_each >= 150
]

discounts = [0.1, 0.2]

discount = np.select(conditions, discounts, default=0)

discounted_revenue = quantity * price_each * (1 - discount)

sum_of_discounted_revenue = discounted_revenue.sum()
plain_revenue = quantity * price_each

mask_pl = country_code == "PL"
only_polish = plain_revenue[mask_pl]

sum_of_only_polish = only_polish.sum()
```
